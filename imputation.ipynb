```python
import pandas as pd

# Read the files
items = pd.read_csv("Data/items.csv")
stores = pd.read_csv("Data/stores.csv")
train = pd.read_csv("Data/train.csv", nrows=10000)

combined_data = pd.merge(train, items, on='item_nbr', how='left')

combined_data = pd.merge(combined_data, stores, on='store_nbr', how='left')

combined_data.head()
```

```python
sale_dat_item = combined_data.groupby(['item_nbr', 'store_nbr', 'date']).agg({'unit_sales': 'sum'}).reset_index()
sale_dat_item.head()
```
